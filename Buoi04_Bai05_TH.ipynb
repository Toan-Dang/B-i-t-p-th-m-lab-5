Chuyển đổi biến diagnosis từ M(ác tính) và B(lành tính) thành 0 và 1

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

<p> tạo 2 tập dữ liệu feature và label

```python
# Scaling the dataset
label = pd.DataFrame(preprocessing.scale(df.iloc[:,1:32]))
label.columns = list(df.iloc[:,1:32].columns)
label['diagnosis'] = df['diagnosis']
# Creating the high dimensional feature space X
feature = label.drop('diagnosis',axis=1)
X = feature.values

#Creating a 2D visualization to visualize the clusters

```

<p>Dùng phương pháp elbow thể hiện gom cụm</p>

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
